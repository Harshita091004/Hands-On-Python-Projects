```python
import requests

query = input(" enter the type of news you are intersted in today")
api = "043754971d7d4c5783cd1f29bf866de1"

url = f"https://newsapi.org/v2/everything?q={query}&from=2025-07-11&sortBy=publishedAt&apiKey={api}"

print(url)
r= requests.get(url)

data = r.json()
articles = data["articles"]

for index, article in enumerate(articles):
    print(index+1 , article["title"] , article["url"])
    print("\n************\n")
```
